```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm  import SVC
from sklearn.metrics import accuracy_score
```

```python
x_unstd = df.drop(columns=['target'])
y = df['target']
```

```python
df.info()
```

```python
df.describe()
```

```python
scaler1 = StandardScaler()
x = scaler1.fit_transform(x_unstd)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
dic = {
    'lr':LogisticRegression(),
    'Nb':GaussianNB(),
    'knn':KNeighborsClassifier()

}
for name,model_temp in dic.items():
    model_temp.fit(x_train,y_train)
    y_train_prediction = model_temp.predict(x_train)
    y_test_prediction = model_temp.predict(x_test)

    train_acc = accuracy_score(y_train_prediction,y_train)
    test_acc = accuracy_score(y_test_prediction,y_test)

    print(f'the traning acc is {train_acc}, and the testing acc is {test_acc}')
```

```python
model1 = SVC()
```

```python
model1.fit(x_train,y_train)
```

```python
y_train_p = model1.predict(x_train)
y_test_p = model1.predict(x_test)
```

```python
acc_of_svctrain = accuracy_score(y_train,y_train_p)
acc_of_svctest = accuracy_score(y_test,y_test_p)
```

```python
print(f'traning acc{acc_of_svctrain}  testing acc{acc_of_svctest}')
```
