```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Hypertension-risk-model-main.csv')
```

```python
df.info()
```

```python
df = df.dropna(axis=0)
```

```python
x_unstd = df.drop(columns=['Risk'])
y = df['Risk']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm  import SVC
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.isnull().sum()
```

```python
x_unstd.head()
```

```python
scaler3 = StandardScaler()
x = scaler3.fit_transform(x_unstd)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
dic = {
    'lr':LogisticRegression(),
    'Nb':GaussianNB(),
    'knn':KNeighborsClassifier()

}
for name,model_temp in dic.items():
    model_temp.fit(x_train,y_train)
    y_train_prediction = model_temp.predict(x_train)
    y_test_prediction = model_temp.predict(x_test)

    train_acc = accuracy_score(y_train_prediction,y_train)
    test_acc = accuracy_score(y_test_prediction,y_test)

    print(f'the traning acc is {train_acc}, and the testing acc is {test_acc}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc =  RandomForestClassifier(
    n_estimators=100,max_depth=9,min_samples_split=7,max_features='sqrt',random_state=42
)
rfc.fit(x_train,y_train)
```

```python
y_train_pred = rfc.predict(x_train)
y_test_pred = rfc.predict(x_test)
```

```python
acc1 = accuracy_score(y_train,y_train_pred)
acc2 = accuracy_score(y_test,y_test_pred)
```

```python
print(f'train acc {acc1} .. test acc {acc2}')
```
