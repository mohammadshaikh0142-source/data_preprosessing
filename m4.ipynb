```python
df['GENDER'] = df['GENDER'].replace({'M':1,'F':0})
```

```python
x_unstd = df.drop(columns=['LUNG_CANCER'])
y = df['LUNG_CANCER']
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm  import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
scaler4 = StandardScaler()
x = scaler4.fit_transform(x_unstd)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
dic = {
    'lr':LogisticRegression(),
    'Nb':GaussianNB(),
    'knn':KNeighborsClassifier()

}
for name,model_temp in dic.items():
    model_temp.fit(x_train,y_train)
    y_train_prediction = model_temp.predict(x_train)
    y_test_prediction = model_temp.predict(x_test)

    train_acc = accuracy_score(y_train_prediction,y_train)
    test_acc = accuracy_score(y_test_prediction,y_test)

    print(f'the traning acc is {train_acc}, and the testing acc is {test_acc}')
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(x_train,y_train)
```

```python
y_train_pred = model_lr.predict(x_train)
y_test_pred = model_lr.predict(x_test)
```

```python
acc1 = accuracy_score(y_train,y_train_pred)
acc2 = accuracy_score(y_test,y_test_pred)
```

```python
print(f'trian acc {acc1} ,, test acc{acc2}')
```
